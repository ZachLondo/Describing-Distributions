## Plotting

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import matplotlib as mpl
import seaborn as sns
```

```python
df = pd.read_csv('https://docs.google.com/spreadsheets/d/1APV3pXiAszS_0mSgkiEt9IUNH-QmyX7KwxSAwuADl6Y/gviz/tq?tqx=out:csv&sheet=medical_data')
df.head()
```

```python
# Analyzing VitD_levels
```

```python
# Plot a histogram with a kde (kernel density estimate)

# Plot and standardize for use with other variables
col = 'VitD_levels'
feature = df[col]
sns.kdeplot(feature)
```

```python
# Add a line for the mean (red)
# Add a line for the median (green)

# Assign values for mean and median
mean = feature.mean()
median = feature.median()

# Replot with lines
fig,ax = plt.subplots(figsize=(10,6))
sns.kdeplot(feature)
ax.axvline(mean, color= 'red')
ax.axvline(median, color='green')
```

```python
# Add a line for for +1 std from the mean (black)
# Add a line for the - 1 std from the mean (black)
# Highlight the range between +1 and =1 std (yellow)

# Assign values for one std above and below mean
std = feature.std()
plus_one_std = mean + std
minus_one_std = mean - std
```

```python
# Replot with better labeling
col = 'VitD_levels'
feature = df[col]
mean = feature.mean()
median = feature.median()
std = feature.std()
plus_one_std = mean + std
minus_one_std = mean - std
fig,ax = plt.subplots(figsize=(10,6))
sns.kdeplot(feature, ax=ax)
ax.axvline(mean, color= 'red', label=f'mean = {mean:,.2f}')
ax.axvline(median, color='green', label=f'median = {median:,.2f}')
# Lines for std's
ax.axvline(plus_one_std, color = 'black',label=f'+1 std = {plus_one_std:,.2f}')
ax.axvline(minus_one_std, color = 'black', label = f'-1 std = {minus_one_std:,.2f}')
# highlight range betweem std's
ax.axvspan(plus_one_std, minus_one_std, color = 'yellow', zorder = 0)
ax.set_title(f'{col}')
ax.legend();
```

## Questions

Is it Discrete or Continuous?

    These variables are all continuous as they can take on a nearly infinite number of values. 

Does it have a skew? If so, which direction (+/-)

    The plot for Total charge has a positive skew.
    The plots for Vitamin D levels and Doctors Visits have no skew

What type of kurtosis does it display? (Mesokurtic, Leptokurtic, Platykurtic)

```python
# Import and run the Pearson Calculation for each
from scipy import stats
kurt = stats.kurtosis(df['VitD_levels'], fisher = False)
kurt
```

```python
kurt = stats.kurtosis(df['Doc_visits'], fisher = False)
kurt
```

```python
kurt = stats.kurtosis(df['TotalCharge'], fisher = False)
kurt
```

Based on the numbers above, I can conclude that the Vitamin D level and Doctors visit distributions are both Mesokurtic.

The distribution for Total Charge is slightly Leptokurtic
